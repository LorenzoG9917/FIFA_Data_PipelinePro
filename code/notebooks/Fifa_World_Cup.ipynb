### 2.4.4 Convertir tipos de datos en las columnas

````python
matches_worldcup.head()
````

````python
#Convertir multiples columnas usando un diccionario
# using dictionary to convert specific columns
convert_dict = {
                'Year': int,
                'Home Team Goals': int,
                'Away Team Goals': int,
                'Attendance':int,
                'Half-time Home Goals':int,
                'Half-time Away Goals':int,
                'RoundID':int,
                'MatchID':int
                }


matches_worldcup = matches_worldcup.astype(convert_dict)
````

````python
matches_worldcup.head()
````

````python
#Convertir columna de datetime
#Formatear cada registro de datetime cuando se adecuado formato
# Create a mapping of date formats and their corresponding format strings
format_mapping = {
    r'%d %b %Y - %H:%M ': 0,
    r'%d %B %Y - %H:%M ': 1,
    r'%d %b %Y %H:%M': 2,
    r'%Y-%m-%d %H:%M': 3,
    r'%B %d, %Y %I:%M %p': 4
}

# Preprocess the data
def preprocess_datetime(datetime_str):
    for format_str, format_id in format_mapping.items():
        try:
            return pd.to_datetime(datetime_str, format=format_str)
        except ValueError:
            continue
    return pd.NaT

# Convert the "Datetime" column using the preprocess function
matches_worldcup['Datetime'] = matches_worldcup['Datetime'].apply(preprocess_datetime)
````
